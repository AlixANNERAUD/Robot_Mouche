```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Dimensions de l'image
largeur = 1280
hauteur = 720

# Création d'une matrice blanche
image = np.ones((hauteur, largeur, 3), dtype=np.uint8) * 255

# Définition des coordonnées de début et de fin de la ligne noire
x_start = np.random.randint(0, largeur)
y_start = 0
x_end = np.random.randint(0, largeur)
y_end = hauteur - 1

# Dessiner la ligne noire
cv2.line(image, (x_start, y_start), (x_end, y_end), (0, 0, 0), thickness=30)

# Ajouter du bruit gaussien à l'image
mean = 0
std_dev = 25
noise = np.random.normal(mean, std_dev, image.shape).astype(np.uint8)
image = image * noise


```

```python

# Affichage de l'image
plt.imshow(image)
plt.show()
```

```python
target = image[hauteur-40 : hauteur-20, :, :]

inverted_target = 255 - target

plt.imshow(inverted_target)
```

```python
mean = inverted_target.mean(axis=0)

plt.bar(np.arange(1280), mean[:, 0], color='red')
```
