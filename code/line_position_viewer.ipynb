```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, clear_output
import time

file_path = 'line_position.bin'

try:
    while True:
        try:
            with open(file_path, 'rb') as file:
                data = file.read(640)
        except FileNotFoundError:
            # Handle the case when the file doesn't exist
            print(f"File not found: {file_path}")
            time.sleep(0.1)
            continue
        array_values = list(data)
        

        # Plot the array
        plt.clf()
        plt.ylim(0, 255)
        plt.plot(array_values)
        display(plt.gcf())
        clear_output(wait=True)

        # Pause for 100ms
        time.sleep(0.1)

except KeyboardInterrupt:
    # Handle KeyboardInterrupt (e.g., stop the loop when interrupting the kernel)
    pass
```
